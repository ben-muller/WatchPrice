```python
import requests
import shutil

def DownLoadImage(url):
    # url = 'https://cdn2.chrono24.com/images/uhren/15239175-pki46i0ev2fjg817j0e2oeq7-ExtraLarge.jpg'
    filename = url.split('/')[-1]
    r = requests.get(url, stream=True)
    if r.status_code == 200:
        with open(f'../Images/{filename}', 'wb') as f:
            r.raw.decode_content = True
            shutil.copyfileobj(r.raw, f)
```

```python
for url in df.iloc[:5]['image.contentUrl'].dropna().values:
    DownLoadImage(url)
```
