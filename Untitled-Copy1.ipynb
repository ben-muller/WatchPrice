```python
import pandas as pd
pd.set_option('display.max_columns', None)
df = pd.read_feather('Data/ListingDB.feather').set_index('index')
searchdf = pd.read_feather('Data/RolexSubmarinerListings_Yearly.feather').set_index('index')
df = df.merge(searchdf[['id','image.contentUrl']], left_index=True,right_on='id', validate='1:1')
df['Image'] = '../Images/'+df['image.contentUrl'].str.extract(r'uhren/(.*?\.jpg)')
df.head()
```

```python
from PIL import Image
import numpy as np
def imtoarray(row):
#     display(row)
    im = Image.open(row['Image'])
    width, height = im.size   # Get dimensions
    left = (width - min(im.size))/2
    top = (height - min(im.size))/2
    right = (width + min(im.size))/2
    bottom = (height + min(im.size))/2
    im = im.crop((left, top, right, bottom))
#     display(im.resize((100,int(im.size[1]/im.size[1]*100))))
    return(np.array(im.resize((100,int(im.size[1]/im.size[1]*100))))/255)

train_images = np.array(df.head(50).apply(lambda row: imtoarray(row), axis=1).values)
train_labels = np.array(df.head(50)['Dial'].astype('category').cat.codes.values)
```

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
```

```python
tf.config.list_physical_devices('GPU')
```

```python
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()

# Normalize pixel values to be between 0 and 1
train_images, test_images = train_images / 255.0, test_images / 255.0
```

```python
train_images.shape
```

```python
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i+50], cmap=plt.cm.binary)
    # The CIFAR labels happen to be arrays, 
    # which is why you need the extra index
    plt.xlabel(train_labels[i+50])
plt.show()
```

```python
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(100, 100, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(4))
model.summary()
```

```python
from datetime import datetime
logdir = "logs/fit/" + datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)


model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

history = model.fit(train_images, train_labels, epochs=20, 
                    validation_data=(test_images, test_labels),
                   callbacks=[tensorboard_callback])
```

```python
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0, 1])
plt.legend(loc='lower right')

test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
```
