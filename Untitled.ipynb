```python
import pandas as pd
def GetWatchDetails(name, price, link):
    url = f'https://www.chrono24.com.au/rolex/{link}'
    try:
        page = pd.read_html(url)
    except Exception as e:
        print(url)
        print(e)
    # df = pd.read_html(url)[0]
    # df

#     display(page[0].set_index(0).transpose())
    # display(pd.read_html(url)[0].pivot(None,0,1))
#     display(page[1].set_index(0).transpose())

#     display(pd.concat([page[0].set_index(0).transpose(),page[1].set_index(0).transpose()], axis=1))
    tempdf = pd.concat([page[0].set_index(0).transpose(),page[1].set_index(0).transpose()], axis=1)
    if 'Others' in tempdf.columns:
        tempdf = tempdf.loc[:,:'Others']
    if 'Functions' in tempdf.columns:
        tempdf = tempdf.loc[:,:'Functions']
#     display(tempdf.columns)
    tempdf['price'] = price
    tempdf['name'] = name
    tempdf['url'] = url
    tempdf.index = [link]
    tempdf = tempdf.groupby(level=0, axis=1).last()
#     display(tempdf.columns.duplicated())
    return(tempdf)
```

```python
import requests
from pandas.io.json import json_normalize  
import json
from tqdm import tqdm_notebook
outdf = pd.DataFrame()
for index in tqdm_notebook(range(100)):
    url = f'https://www.chrono24.com.au/rolex/index-{index}.htm?pageSize=120&sortorder=5&resultview=block'
    r = requests.get(url)
    if r.url != url:
        print(url)
        print('done')
        break
    j = r.text.split('<script type="application/ld+json">\n')[1].split('\n   </script>')[0].replace('\n','').replace('\t','')
    j = json.loads(j) 
    df = json_normalize(j['@graph'][1]['offers']) 
    outdf = outdf.append(df, sort=False)
```

```python
outdf2 = pd.DataFrame()

for index, row in tqdm_notebook(outdf.head(1000).iterrows(), total=len(outdf.head(1000))):
#     outdf2 = pd.concat([outdf2,GetWatchDetails(row['name'], row['price'], row['link'])], axis=1, join='inner')
    outdf2 = outdf2.append(GetWatchDetails(row['name'], row['price'], row['link']), sort=False)
display(outdf2)
```

```python
outdf2.to_csv('watches.csv')
```
